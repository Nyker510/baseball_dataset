```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import calstats as cstat
import seaborn as sns

%matplotlib inline

import matplotlib
font = {'family':'IPAGothic'}
matplotlib.rc('font', **font)
```

```python
cwd = os.getcwd()
os.chdir('data')
os.chdir('hitter')
```

```python
os.getcwd()
```

```python
data = []
for p in os.listdir():
    if 'csv' not in p:
        continue
    print(p)
    df = pd.read_csv(p,index_col=0)
    df['date'] = pd.to_datetime(df['date'])
    df['pa'] = 1
    df = df.groupby('date').sum().cumsum()
    df = cstat.calclate_ratio_index(df)
    data.append({
            'name':p,
            'dataframe':df
                })
```

```python
os.chdir(cwd)
ls_stats = ['at_bats', 'double', 'bant', 'triple', 'hr', 'single', 'k', 'sac_fly',
       'bb', 'pa', 'hit', 'bases', 'ab', 'AVG', 'OBP', 'KP', 'BP', 'SLG',
       'OPS', 'IsoD', 'IsoP', 'BABIP', 'XR', 'XR27']
for stats in ls_stats:
    data.sort(key=lambda d:d['dataframe'][stats].tail(1).values[0])
    df_st = pd.DataFrame()
    for i,d in enumerate(data[-5::][::-1]):
        print(d['name'])
        df = d['dataframe']
        name = d['name'][5:-4]
        df_st[name] = df[stats]
    #     plt.plot(df.index,df[stats],label=d['name'])
    df_st = df_st.fillna(method='ffill')
    
    plt.figure(figsize=(8,8))
    df_st.plot()
    plt.tight_layout()
    plt.savefig('figures/{}_cumsum.png'.format(stats),dpi=100)
```

```python
os.chdir(cwd)

l_stats = ['bb','k','XR','hit','hr','IsoD']
for stats in l_stats:
    data.sort(key=lambda d:d['dataframe'][stats].tail(1).values[0])
    df_gather = pd.DataFrame()
    for i,d in enumerate(data[-5::]):
        df = d['dataframe']
        name = d['name'][5:-4]
        df_gather[name] = df[stats]
    df_gather.fillna(method='bfill').to_csv('{0}_cumsum.csv'.format(stats))
```
